```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal
import math
pi = tf.constant(math.pi)
M = 256
```

```python
def rmse(predictions,targets):
  return (np.sqrt((np.subtract(predictions,targets) ** 2).mean()))   


message = np.arange(1,M)
rmse_uw = np.load('rmse_uw_200mW.npy')
rmse_w = np.load('rmse_w_200mW.npy')
rmse_Tb = np.load('rmse_Tb_200mW.npy')
#print(rmse_w)
plt.plot(message, rmse_uw, '-o');
plt.plot(message, rmse_w, '-*');
plt.plot(message, rmse_Tb, '-+');
plt.xlabel('Distance')
plt.ylabel('RMSE')
#plt.ylim([0, 20])
#plt.ylabel('log2 Norm RMSE deviation')
#plt.legend(['Weighted ^1', 'Weighted ^2'])
plt.legend(['SPL', 'Weighted SPL','Baseline'])
plt.rcParams.update({'font.size': 15})
plt.ylim([0, 100])

print(np.sum(rmse_uw))
print(np.sum(rmse_w))
print(np.sum(rmse_Tb))
#print(np.sum(rmse_uwA))
#print(np.sum(rmse_wA))
```
